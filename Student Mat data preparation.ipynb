```python
import pandas as pd
```

```python
data_m = pd.read_csv('student-mat.csv',sep=';')
data_p = pd.read_csv('student-por.csv',sep=';')
data_t = pd.merge(data_p,data_m,how='left')
```

```python
data_t.head()
```

```python
data_t.describe().T
```

```python
to_drop = ['school','age','school','famsize','Pstatus','reason','guardian','romantic','famrel','Dalc','Walc']
data_mat = data_t.drop(to_drop, axis=1)
```

```python
   
def change_job(job):
    if job == 'at_home':
        return 0
    elif job == 'other':
        return 1
    elif job == 'teacher':
        return 2
    elif job == 'services':
        return 3
    else:
        return 4
```

```python
data_mat['Mjob'] = data_mat['Mjob'].apply(change_job)
data_mat['Fjob'] = data_mat['Fjob'].apply(change_job)
```

```python
data_mat.head()
```

```python
 data_mat['address'] = data_mat['address'].apply(lambda x: 1 if x =='U' else 0)
 data_mat['sex'] = data_mat['sex'].apply(lambda x: 1 if x =='M' else 0)
 data_mat['address'] = data_mat['address'].apply(lambda x: 1 if x == 'U' else 0)


# YES/NO change


 data_mat['schoolsup'] = data_mat['schoolsup'].apply(lambda x: 1 if x =='yes' else 0)
 data_mat['famsup'] = data_mat['famsup'].apply(lambda x: 1 if x =='yes' else 0)
 data_mat['paid'] = data_mat['paid'].apply(lambda x: 1 if x =='yes' else 0)
 data_mat['activities'] = data_mat['activities'].apply(lambda x: 1 if x =='yes' else 0)
 data_mat['nursery'] = data_mat['nursery'].apply(lambda x: 1 if x =='yes' else 0)
 data_mat['higher'] = data_mat['higher'].apply(lambda x: 1 if x =='yes' else 0)
 data_mat['internet'] = data_mat['internet'].apply(lambda x: 1 if x =='yes' else 0)
```
